## Geomagic VS UdMC

### Mean absolute error (MAE)

```python
udmc - geom
```

```python
(udmc - geom).abs().mean()
```

### Mean relative error (MRE)

```python
(udmc - geom) / geom.mean()
```

```python
((udmc - geom) / geom.mean()).plot(kind='bar')
```

```python
((udmc - geom) / geom.mean()).abs().mean()
```
